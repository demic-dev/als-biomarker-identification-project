```python
from modules.RNA import RNASeq
import glob

sequencing = RNASeq(verbose=True)
sequencing.load_bulk_data(glob.glob("data/*.txt"))
sequencing.load_annotations("data/GSE124439_family.xml")
```

```python
std = sequencing.get_std_genes()
median = sequencing.get_median_genes()
mean = sequencing.get_mean_genes()
```

```python
import matplotlib.pyplot as plt
```

```python
# median histogram
median.plot.hist()
plt.show()
# std histogram
std.plot.hist()
plt.show()
# mean histogram
mean.plot.hist()
plt.show()
```

```python
sequencing.get_sample_annotations()["sample group"].value_counts().plot(kind='bar', rot=45)
plt.show()
sequencing.get_sample_annotations()["cns subregion"].value_counts().plot(kind='bar', rot=45)
plt.show()
sequencing.get_sample_annotations()["tissue type"].value_counts().plot(kind='bar', rot=45)
plt.show()
```
